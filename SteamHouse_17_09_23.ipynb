```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Replace the file path with the path to your Excel file
file_path = '/Users/mdabusufian/Downloads/New_Dataset_Business.xlsx'

# Read the data from the Excel file
df = pd.read_excel(file_path, engine='openpyxl')

# Fill missing values with the mean of the respective columns
df['Coal_Consumption'].fillna(df['Coal_Consumption'].mean(), inplace=True)

# Perform the One-Sample Kolmogorov-Smirnov Test for normality
coal_consumption = df['Coal_Consumption']
mean, std = coal_consumption.mean(), coal_consumption.std()
D, p_value = stats.kstest(coal_consumption, 'norm', args=(mean, std))

print(f"One-Sample Kolmogorov-Smirnov Test for 'Coal_Consumption_By':")
print(f"D-statistic: {D}")
print(f"p-value: {p_value}")

if p_value > 0.05:
    print("Fail to reject the null hypothesis. The sample follows a normal distribution.")
else:
    print("Reject the null hypothesis. The sample does not follow a normal distribution.")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Replace the file path with the path to your Excel file
file_path = '/Users/mdabusufian/Downloads/New_Dataset_Business.xlsx'

# Read the data from the Excel file
df = pd.read_excel(file_path, engine='openpyxl')

# Fill missing values with the mean of the respective columns
df['Electricity _Consumption'].fillna(df['Electricity _Consumption'].mean(), inplace=True)

# Extract the 'Electricity_Consumption' column
electricity_consumption = df['Electricity _Consumption']

# Calculate the mean and standard deviation of the 'Electricity_Consumption' column
mean = np.mean(electricity_consumption)
std_dev = np.std(electricity_consumption)

# Define the cumulative distribution function (CDF) for a normal distribution with the same mean and standard deviation
def normal_cdf(x):
    return stats.norm.cdf(x, loc=mean, scale=std_dev)

# Perform the One-Sample Kolmogorov-Smirnov Test
ks_statistic, p_value = stats.kstest(electricity_consumption, normal_cdf)

print(f"Kolmogorov-Smirnov statistic: {ks_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```
